```python
from meta_planning.parsers import parse_trajectory, parse_model
from meta_planning import LearningTask
```

```python
M = parse_model('../src/meta_planning/dataset/blocks/ref_domain')
```

```python
t1 = parse_trajectory('../src/meta_planning/dataset/blocks/trajectory-01', M.predicates)
t2 = parse_trajectory('../src/meta_planning/dataset/blocks/trajectory-02', M.predicates)
trajectories = [t1, t2]
```

```python
O = [t.observe(1, action_observability = 1, keep_every_state=True) for t in trajectories]
```

```python
o = O[0]
o.all_states_observed
```

```python
lt = LearningTask(M,O)
```

```python
print(lt.compiled_model)
```

```python
solution = lt.learn()
```

```python
print(solution.learned_model)
```

```python
solution.solution_plan
```
