```python
from ltlf2dfa.parser.ltlf import LTLfParser
from graphviz import Source
import pythomata
import pydot
from pythomata import SimpleDFA
```

```python
parser = LTLfParser()
formula_str = "F(X(a & X(F(b))))"
formula = parser(formula_str)       # returns an LTLfFormula

print(formula)
```

```python
dfa = formula.to_dfa()
print(dfa)                          # prints the DFA in DOT format
```

```python
dfa
```

```python
s = Source(dfa)
s.view()
```

```python
graphs = pydot.graph_from_dot_data(dfa)
graph = graphs[0]
```

```python
for node in graph.get_nodes():
    print(node.get_name(), node.get_shape())
```

```python
for edge in graph.get_edges():
    print(edge.get_source(), edge.get_destination(), edge.get_label())
```
